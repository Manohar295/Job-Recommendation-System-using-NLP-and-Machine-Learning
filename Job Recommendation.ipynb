```python
df = df.drop(['company','location','experience_years','salary'],axis=1)
df
```

```python
df["skills_required"].value_counts()
```

```python
# converting the text data to feature vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(df["skills_required"])
```

```python
vectorizer.get_feature_names_out()
```

```python
pd.DataFrame(vectorizer.fit_transform(df["skills_required"]).toarray(),columns=vectorizer.get_feature_names_out())
```

```python
# getting the similarity scores using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
df["job_title"].tolist()
```

```python
skills = input("Enter your skills: ").split(",")
skills = [skill.strip() for skill in skills]

user_input = " ".join(skills)

user_vector = vectorizer.transform([user_input])

similarity_scores = cosine_similarity(user_vector, feature_vectors)

# flatten the array (2D → 1D)
similarity_scores = similarity_scores.flatten()

# get top 5 indices (highest similarity)
top_5_indices = similarity_scores.argsort()[-10:][::-1]

print("\nTop 5 Job Recommendations:\n")

for i in top_5_indices:
    print(f"Job: {df['job_title'].iloc[i]} |  Match Score: {(similarity_scores[i])}")
```
